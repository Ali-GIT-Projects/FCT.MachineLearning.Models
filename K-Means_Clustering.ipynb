```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
ds = pd.read_csv('Mall_Customers.csv')
x= ds.iloc[:, 3: ].values
print(x)
```

```python
plt.scatter(x[:,0], x[:,1])
plt.title('Income vs Spending')

```

```python
km = KMeans(n_clusters=5)
```

```python
y_pred = km.fit_predict(x)
y_pred
```

```python
x = np.column_stack((x, y_pred))
x
```

```python
km.cluster_centers_
```

```python
x1 = x[x[:,2]==0]
x2 = x[x[:,2]==1]
x3 = x[x[:,2]==2]
x4 = x[x[:,2]==3]
x5 = x[x[:,2]==4]

plt.scatter(x1[:,0],x1[:,1], color = 'green', label = 'Cluster 1')
plt.scatter(x2[:,0],x2[:,1], color = 'red', label = 'Cluster 2')
plt.scatter(x3[:,0],x3[:,1], color = 'blue', label = 'Cluster 3')
plt.scatter(x4[:,0],x4[:,1], color = 'cyan', label = 'Cluster 4')
plt.scatter(x5[:,0],x5[:,1], color = 'yellow', label = 'Cluster 5')

plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:, 1], s=100, marker= '*', color ='purple', label= 'centroids')

plt.title('Clustors of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

```python
sse = []
for i in range(1, 11):
    km = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    km.fit(x)
    sse.append(km.inertia_)
print(sse)
```

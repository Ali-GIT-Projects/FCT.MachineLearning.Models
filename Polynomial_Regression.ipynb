```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Position_Salaries1.csv')
x = df.iloc[:, 1:2].values
y = df.iloc[:, 2].values
print(x)
print(y)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
pr = PolynomialFeatures(degree=4)
x_poly = pr.fit_transform(x)
lr2 = LinearRegression()
lr2.fit(x_poly, y)
```

```python
plt.scatter(x, y, color='orange')
plt.plot(x, lr.predict(x), color='green')
plt.title('Linear Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(x, y, color='orange')
plt.plot(x, lr2.predict(pr.transform(x)), color='green')
plt.title('Polynomial Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
lr.predict([[5.5]])
```

```python
lr2.predict(pr.fit_transform([[5.5]]))
```

CLEARLY HERE THE POLYNOMIAL REGRESSION PREDICTION IS FAR MORE ACCURATE AND REALISTIC BASED ON THE EXCEL DATA OF SALARIES!!!!!
